```
import numpy as np
import scipy
import scipy.stats
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme(style="darkgrid")
```

```
def target(x):
  temp= (np.sin(x)*np.sin(2*x))**2* scipy.stats.norm.pdf(x)
  return temp

def MH(T=10**4, alpha=1):
  x0=3.141
  sim=np.zeros(T+1)
  sim[0]=x0
  for i in range(1,T+1):
    y= scipy.stats.uniform.rvs(size=1, loc= sim[i-1]-alpha, scale= 2*alpha)
    rho=min(1,target(y)/target(sim[i-1]))
    u= scipy.stats.uniform.rvs(size=1)
    if u <= rho:
      sim[i]=y 
    else: 
      sim[i]=sim[i-1]

  return sim
```

```
x=np.linspace(-3,3,1000)

#Funcion sin normalizar


plt.plot( np.linspace(-3,4,1000) ,target(x))
plt.title('Funcion objetivo sin normalizar '+r'$\tilde{\pi}(x)$')
plt.xlabel('x')
plt.ylabel(r'$\tilde{\pi}(x)$')
plt.savefig('1.png', dpi=500)
```

```
#---- alpha=0.1

alpha=0.1
x=MH(T=10**5,alpha=alpha)
sns.histplot(x)
plt.xlabel('x')
plt.ylabel('Conteos')
plt.title('Histograma de x usando '+r'$\alpha$='+str(alpha))
plt.savefig('2.png', dpi=500)
```

```
#---- alpha=0.2

alpha=0.2
x=MH(T=10**5,alpha=alpha)
sns.histplot(x)
plt.xlabel('x')
plt.ylabel('Conteos')
plt.title('Histograma de x usando '+r'$\alpha$='+str(alpha))
plt.savefig('3.png', dpi=500)
```
